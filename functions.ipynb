```python
inputs = list(range(10))
def const(x):
    return 4

def linear(x):
    return x * 2 + 2

def square(x):
    return x * x + x + 2

def exponential(x):
    return pow(2, x) / 1000000

plot_functions(inputs, [const, linear, square])
```

```python
def powerful(x):
    return pow(x, 5) / 10000

n = 10 # square is more with n=10, but try setting n to bigger values!

plot_functions(n, [square, powerful])
```

```python
def exponential(x):
    return pow(2, x) / 1000000

n = 10 # exponential is less with n=10, but try setting n to bigger values!
plot_functions(n, [powerful, exponential])
```

```python
import math

def log_amplified(x):
    return math.log2(x + 1) * 10


n = 10 # log is increasing veeery slow, try bigger n
plot_functions(n, [linear, log_amplified])
```

```python
# Ratio of functions with the same growth rate will tend to a constant

def another_square(x):
    return 3 * x * x + 2 * x - 3

def square_ratio(x):
    return square(x) / another_square(x)

n = 20
plot_functions(n, [square_ratio])
```

```python
def another_exponential(x):
    return pow(2, x) + x * x

def exponential_ratio(x):
    return exponential(x) / another_exponential(x)

n = 20
plot_functions(n, [exponential_ratio])
```
